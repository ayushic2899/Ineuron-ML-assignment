# DATA VISUALIZATION

### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

Reduce dimension of data as we want 3 features to extract
(PCA for Data Visualization For a lot of machine learning applications it helps to be able to visualize your data.)

```python
X_pca = PCA(n_components = 3).fit_transform(iris.data)
X_pca 
```

gives prediction or target  values

```python
y = iris.target 
```

plot 3D plot with elevation and azimuthal angle given

```python
fig = plt.figure(1,figsize = (11,7)) 
x = Axes3D(fig, elev=-150, azim=120)  
x.scatter(X_pca[:,0],X_pca[:,1],X_pca[:,2], c = y, cmap = plt.cm.rainbow_r)
x.set_title("iris dataset in 3d plot")
x.set_xlabel("x data")
x.set_ylabel("y data")
x.set_zlabel("z data")
plt.show()
```
